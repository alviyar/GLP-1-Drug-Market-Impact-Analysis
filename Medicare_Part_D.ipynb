```python
glp1_drugs = ['Ozempic', 'Wegovy', 'Mounjaro', 'Zepbound', 'Trulicity', 'Rybelsus', 'Saxenda', 'Victoza']

df_glp1 = df[df['Brnd_Name'].isin(glp1_drugs)]
df_glp1[['Tot_Benes_2019','Tot_Benes_2020','Tot_Benes_2021','Tot_Benes_2022','Tot_Benes_2023']].isna().sum()
```

```python
df_glp1.set_index('Brnd_Name')[['Tot_Benes_2019','Tot_Benes_2020','Tot_Benes_2021','Tot_Benes_2022','Tot_Benes_2023']]
```

```python
df = df[df['Brnd_Name'].isin(glp1_drugs)].reset_index(drop=True)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.fillna(0)
```

```python
df["Spending_Growth_%"] = (
    (df["Tot_Spndng_2023"] - df["Tot_Spndng_2019"])
    / (df["Tot_Spndng_2019"] + 1)
) * 100
```

```python
df.head()
```

```python
df["Beneficiary_Growth_%"] = (
    (df["Tot_Benes_2023"] - df["Tot_Benes_2019"])
    / (df["Tot_Benes_2019"] + 1)
) * 100
```

```python
df["Claim_Growth_%"] = (
    (df["Tot_Clms_2023"] - df["Tot_Clms_2019"])
    / (df["Tot_Clms_2019"] + 1)
) * 100
```

```python
df["Price_Change_%"] = (
    (df["Avg_Spnd_Per_Clm_2023"] - df["Avg_Spnd_Per_Clm_2019"])
    / (df["Avg_Spnd_Per_Clm_2019"] + 1)
) * 100
```

```python
top10 = df["Tot_Spndng_2023"].nlargest(10).min()

df["Top10_Drug"] = (
    df["Tot_Spndng_2023"] >= top10
).astype(int)
```

```python
df["New_Drug"] = (
    (df["Tot_Spndng_2019"] == 0) &
    (df["Tot_Spndng_2023"] > 0)
).astype(int)
```

```python
spending_cols = [
    "Tot_Spndng_2019",
    "Tot_Spndng_2020",
    "Tot_Spndng_2021",
    "Tot_Spndng_2022",
    "Tot_Spndng_2023"
]

df["Avg_5Y_Spending"] = df[spending_cols].mean(axis=1)
```

```python
df["Spending_Volatility"] = df[spending_cols].std(axis=1)
```

```python
df["Avg_5Y_Spending"] = df[
    ["Tot_Spndng_2019","Tot_Spndng_2020","Tot_Spndng_2021","Tot_Spndng_2022","Tot_Spndng_2023"]
].mean(axis=1)
```

```python
top10 = df.nlargest(10, "Tot_Spndng_2023")

plt.figure(figsize=(10,5))
plt.bar(top10["Brnd_Name"], top10["Tot_Spndng_2023"])
plt.xticks(rotation=45)
plt.title("Top 10 Drugs by Spending (2023)")
plt.ylabel("Total Spending")
plt.tight_layout()
plt.show()
```

```python
top10 = df.nlargest(10, "Tot_Spndng_2023")

years = ["2019","2020","2021","2022","2023"]

plt.figure(figsize=(12,6))

for _, row in top10.iterrows():
    spending = [
        row["Tot_Spndng_2019"],
        row["Tot_Spndng_2020"],
        row["Tot_Spndng_2021"],
        row["Tot_Spndng_2022"],
        row["Tot_Spndng_2023"]
    ]

    plt.plot(years, spending, marker='o', label=row["Brnd_Name"])

plt.title("Top 10 Drugs Spending Trend (2019-2023)")
plt.xlabel("Year")
plt.ylabel("Total Spending")
plt.legend(bbox_to_anchor=(1.05,1), loc="upper left")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Tot_Spndng_2023"], bins=30)
plt.title("Distribution of Drug Spending (2023)")
plt.xlabel("Spending")
plt.ylabel("Frequency")
plt.show()
```
